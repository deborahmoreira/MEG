# Tópico 2 - Lendo e imprimindo valores


Em muitas situações, queremos poder fornecer uma informação para o computador, para que ele possa processar a informação, e depois, exibir o resultado em sua tela. Para fazer isso, podemos utilizar duas funções básicas:

* `print()`: do inglês, significa imprimir, ou seja, vamos imprimir informações na tela do computador

## Exemplo 2 - Escrevendo e imprimindo seu nome

Execute o código abaixo e veja o que acontece.

````python
nome = input("Qual é o seu nome?")
print('Olá, ' + nome + '!')
print(f'Olá {nome}!')
````

Nesse código, o computador fica aguardando a entrada de informações. Isso é feito com a função `input`. Quando há a entrada dessa informação, ela é guardada na variável `nome`.

## Exercício 3 - Calculando a força de atração entre a terra e a lua

Pode-se dizer que a Terra atrai os corpos, independentemente do ponto do planeta em que se está. Isso se deve à força da gravidade exercida pela Terra sobre todos os corpos que estejam relativamente próximos de sua superfície.

Os corpos também se atraem com essa força da gravidade, ou seja, se são atraídos pela Terra, eles também têm gravidade e atraem outros corpos, que, por sua vez, também os atraem (Terceira Lei de Newton). Surgiu, assim, a ideia de gravitação universal.

Sejam duas massas m1 e m2, em que d é a distância entre seus centros.

gravitacao-universal.jpg

A formula para calcular a força atrativa entre duas massas é:

$F = G \cdot \frac{m_1 \cdot m_2}{d^{2}}$

Onde G é denominado constante da gravitação universal, sendo seu valor expresso, no Sistema Internacional, por: G = 6,67.10-11 N.m2.Kg2.
m1 e m2 são as massas dos dois corpos. A Terra possui  $5.9722 * 10^24$ kg e a Lua possui $7,35 * 10^22$ kg. d² é o quadrado da distância que os separa, que no caso é 384,400 km.

# Tópico 5 - Estruturas condicionais

As estruturas condicionais são utilizadas quando queremos que nosso programa tome caminhos diferentes, dependendo das informações que o usuário digitar, por exemplo. A figura abaixo mostra um exemplo de uma estrutura condicional:

condicoes.png



Para utilizar uma estrutura condicional no Python, utilizamos os comandos `if`(do inglês, se) e `else``(do inglês, se não). Abaixo, temos um exemplo de código básico:

```
if(condição):
   faça uma coisa
else:
  faça outra coisa
```

Se a minha condição for atendida, o programa irá fazer uma coisa, se não, ele irá fazer outra coisa diferente.


## Exemplo 6: Determinando se o dia está quente ou não

Nessa primeira situação, vamos considerar que uma temperatura acima de 30°C determina que um dia é quente, e qualquer valor abaixo disso, determina que um dia é frio. Observe o código abaixo e mude os valores da variável temperatura para ver como ele se comporta

````python
temperatura = 32
if(temperatura > 30):
  print('Hoje o dia está quente!')
else:
  print('Hoje o dia está frio!')
````

Às vezes nós queremos verificar várias condições diferentes. Por exemplo, quero fazer um programa que diga as seguintes informações de um dia com base em sua temperatura:

* Se a temperatura estiver acima de 30°C, o dia está muito quente
* Se a temperatura estiver entre 20°C e 30°C, o dia está agradável
* Se a temperatura estiver abaixo de 20°C, o dia está muito frio

Para isso, nós podemos acrescentar uma outra estrutura do python, o elif (do inglês, se não se). Veja a estrutura básica abaixo e depois analise o código, trocando os valores da variável temperatura para ver como ele se comporta:

```
if(condição 1):
  faça uma coisa
elif(condição 2):
  faça outra coisa
else:
  faça outra coisa diferente
```




````python
temperatura = 25
if(temperatura > 25):
  print('Hoje o dia está muito quente!')
elif(temperatura >= 20 and temperatura <= 30):
  print('Hoje o dia está agradável')
else:
  print('Hoje o dia está muito frio!')
````
